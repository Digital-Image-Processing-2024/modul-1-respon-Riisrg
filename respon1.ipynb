<h2>Menampilkan Gambar</h2>
<p>Membaca gambar menggunakan fungsi imread dari dan menampilkan gambar menggunakan fungsi imshow dari library matplotlib</p>

```python
import cv2
import matplotlib.pyplot as plt

gambar = cv2.imread("respon.jpg")

gambar_rgb = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)


channel_r, channel_g, channel_b = cv2.split(gambar_rgb)

# Tampilkan gambar dan histogram untuk masing-masing channel
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Tampilkan gambar
axes[0, 0].imshow(gambar_rgb)
axes[0, 0].set_title('Citra RGB')

# Tampilkan histogram untuk masing-masing channel
axes[1, 0].hist(channel_r.flatten(), bins=256, color='red', alpha=0.7)
axes[1, 0].set_title('Histogram R')

axes[1, 1].hist(channel_g.flatten(), bins=256, color='green', alpha=0.7)
axes[1, 1].set_title('Histogram G')

axes[1, 2].hist(channel_b.flatten(), bins=256, color='blue', alpha=0.7)
axes[1, 2].set_title('Histogram B')

# Tampilkan gambar untuk masing-masing channel
axes[0, 1].imshow(channel_r, cmap='Reds')
axes[0, 1].set_title('Channel R')

axes[0, 2].imshow(channel_g, cmap='Greens')
axes[0, 2].set_title('Channel G')

axes[0, 3].imshow(channel_b, cmap='Blues')
axes[0, 3].set_title('Channel B')

plt.tight_layout()
plt.show()
```
