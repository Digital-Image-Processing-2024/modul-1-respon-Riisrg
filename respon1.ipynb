<h2>Import Library Yang Diperlukan</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
```

<h2>Menampilkan Gambar RGB</h2>
<p>Membaca gambar menggunakan OpenCV, ubah menjadi format RGB, dan kemudian pisahkan channel warna merah, hijau, dan biru dari gambar tersebut. Setelah itu, menggunakan Matplotlib, tampilkan gambar asli dan masing-masing channel warna secara terpisah dalam empat subplot berbeda dengan judul yang sesuai.</p>

```python

gambar = cv.imread("respon.jpg")

gambar_rgb = cv.cvtColor(gambar, cv.COLOR_BGR2RGB)


def pisah_warna(gambar):
    channel_r = gambar[:, :, 0]
    channel_g = gambar[:, :, 1]
    channel_b = gambar[:, :, 2]
    return channel_r, channel_g, channel_b

red, green, blue = pisah_warna(gambar)

kanvas, kordinat = plt.subplots(1, 4, figsize=(10, 5))

kordinat[0].imshow(gambar_rgb)
kordinat[0].set_title('Citra RGB')

kordinat[1].imshow(red, cmap='Reds')
kordinat[1].set_title('Citra RED')

kordinat[2].imshow(green, cmap='Greens')
kordinat[2].set_title('Citra GREEN')

kordinat[3].imshow(blue, cmap='Blues')
kordinat[3].set_title('Citra BLUE')

plt.tight_layout()
plt.show()
```

<h2>Menampilkan Histogram</h2>
<p>Berikut merupakan program untuk membuat tiga histogram yang menampilkan distribusi intensitas warna merah, hijau, dan biru dari gambar diatas. Dengan menggunakan subplots dengan satu baris dan tiga kolom, setiap histogram dibuat menggunakan fungsi hist() dengan data dari salinan rata-rata warna merah, hijau, dan biru (red, green, dan blue) yang telah di-flatten.</p>

```python
kanvass, koor = plt.subplots(1,3, figsize=(10,5))

koor[0].hist(red.flatten(), bins=256, color='red', alpha=0.7)
koor[0].set_title('Histogram R')

koor[1].hist(green.flatten(), bins=256, color='green', alpha=0.7)
koor[1].set_title('Histogram G')

koor[2].hist(blue.flatten(), bins=256, color='blue', alpha=0.7)
koor[2].set_title('Histogram B')


plt.tight_layout()
plt.show()

```
